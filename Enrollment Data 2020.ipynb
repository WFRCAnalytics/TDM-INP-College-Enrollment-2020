```python
import pandas as pd
import os
import numpy as np
```

```python
working_directory = os.getcwd()
print(working_directory)
```

```python
# Read 2020 csv as pandas dataframe object
results_folder = os.path.join(os.getcwd(), 'data')
csv_enrol = os.path.join(results_folder, 'headcount.csv')
df_enrol = pd.read_csv(csv_enrol)
```

```python
# Show the first 5 rows of a table
df_enrol.head()
```

```python
df_enrol.groupby('Institution').agg({'Headcount': [np.size, np.sum]})
```

```python
df_enrol.groupby(['Institution','Site']).agg({'Headcount': [np.size, np.sum]})
```

```python
df_enrol.groupby(['Institution','Site']).agg({'Headcount': [np.size, np.sum]}).sum()
```

```python
df_enrol[(df_enrol['Institution'] == 'UU') & (df_enrol['Site'] == 'Main Campus')].groupby(['Institution','Site','ZIP_Code']).agg({'Headcount': [np.size, np.sum]})
```
